```python
import math
import cPickle as pickle
import nltk

from testing_util import sample_sets
```

```python
stopwords = nltk.corpus.stopwords.words('english')
stemmer = nltk.PorterStemmer()

def get_terms(t):
    """
    This is used to get the relevant items out of text. 
    """
    tokens = nltk.word_tokenize(t)
    return [stemmer.stem(w.lower()) for w in tokens if len(w)>3 and (w not in stopwords)]
```

```python
terms = set([t for s in sample_sets for t in get_terms(s[1])])
```

```python
sample_terms = [list(set(get_terms(s[1]))) for s in sample_sets]
```

```python
tdf = nltk.FreqDist(t for l in sample_terms for t in l)
```

```python
def get_tdfs(samples, termer):
    sample_terms = [list(set(termer(s[1]))) for s in sample_sets]
    tdf = nltk.FreqDist(t for l in sample_terms for t in l)
    return tdf
```

```python
def get_cat_tfidfs(samples, termer, categories):
    tdf = get_tdfs(samples, termer)
    
    cat_tfidfs={}
    for c in categories:
        tfidfs={}
        fd = nltk.FreqDist([t for s in sample_sets for t in get_terms(s[1]) if s[0]==str(c)])

        for t in fd.keys():
            tfidfs[t]=fd.freq(t)/idf[t]

        cat_tfidfs[str(c)]=tfidfs
```

```python
idf = {}
for t in terms:
    idf[t]=math.log(float(len(sample_sets))/tdf[t])
```

```python
cat_tfidfs={}

for i in range(1,8):
    tfidfs={}
    fd = nltk.FreqDist([t for s in sample_sets for t in get_terms(s[1]) if s[0]==str(i)])
    
    for t in fd.keys():
        tfidfs[t]=fd.freq(t)/idf[t]
    
    cat_tfidfs[str(i)]=tfidfs
```
