```python
import re
import random
from os import path

import nltk

import testing_util as util
import term_scoring
from testing_util import sample_sets, final_sets

import tfidf_class

from nltk.classify import SklearnClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier, RadiusNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
```

```python
#classifier = nltk.classify.NaiveBayesClassifier
#classifier = SklearnClassifier(SVC(), sparse=False)
classifier = SklearnClassifier(KNeighborsClassifier(n_neighbors=1, weights='uniform'))
#classifier = SklearnClassifier(RandomForestClassifier(), sparse=False)
```

```python
stopwords = nltk.corpus.stopwords.words('english')
stemmer = nltk.PorterStemmer()

def get_terms(t):
    """
    This is used to get the relevant items out of text. 
    """
    tokens = nltk.word_tokenize(t)
    return [stemmer.stem(w.lower()) for w in tokens if len(w)>3 and (w not in stopwords)]
```

```python
def category_term_scorer(sample_list, nterms=100):
    """
    takes a list of tuples in format (category,text) and creates scores for each term
    for its relevance to each category
    """
    
    categories = [str(x) for x in range(1,8)]
    idfs = tfidf_class.get_cat_idfs(sample_list, get_terms, categories)
    
    for k,v in idfs.iteritems():
        idfs[k]=v[0:nterms]
    
    
    scores={}
    
    for cat, termlist in idfs.iteritems():
        #print cat
        for term in termlist:
            #print term
            if term[0] not in scores:
                scores[term[0]]={c: 0 for c in categories}
            scores[term[0]][cat]=term[1]
            
    return scores
```

```python
feature_term_score = term_scoring.TermScoreBagger(scorer=category_term_scorer)
```

```python
extractor = util.FeatureExtractor()
extractor.add_extractor(feature_term_score, trained=True)
```

```python
for i in range(0,10):
    sets = util.get_folds(sample_sets, folds=10)
    train1 = [s for f in sets[0:6] for s in f]
    train2 = [s for f in sets[6:9] for s in f]
    test = [s for s in sets[9]]
    extractor.train_extractors(train1)
    cl = extractor.get_classifier(sample_sets, classifier=classifier, retrain=False)
    test_features = util.make_feature(extractor, test)
    print nltk.classify.accuracy(cl, test_features)
    results = [cl.classify(i[0]) for i in test_features]
    reference = [i[1] for i in test_features]
    cm = nltk.ConfusionMatrix(reference, results)
    print(cm.pp(sort_by_count=True, show_percents=True, truncate=9))
```
